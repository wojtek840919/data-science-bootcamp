```python
import pandas as pd
import numpy as np
```

```python
def fetch_financial_data(company='AMZN'):
  import pandas_datareader.data as web
  return web.DataReader(name=company, data_source='stooq')

df=fetch_financial_data('UBER')
df.head()
```

```python
df.to_csv('uber.csv')
```

```python
df_nov = df[(df.index.month==11)&(df.index.year==2019)]
df_nov
```

```python
df_nov.to_csv('uber_nov.csv')
```

```python
new_df=pd.read_csv('uber_nov.csv', index_col=0)
new_df
```

```python
df_nov.to_excel('uber_nov.xlsx')
```
