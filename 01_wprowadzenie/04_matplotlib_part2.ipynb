```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
img = mpimg.imread('office.jpg')
img
```

```python
img.shape
```

```python
plt.imshow(img)
_=plt.axis('Off')
```

```python
fig = plt.figure(figsize=(12,10))
plt.subplot(221)
plt.imshow(mpimg.imread('office.jpg'))
plt.subplot(222)
plt.imshow(mpimg.imread('business.jpg'))
plt.subplot(223)
plt.imshow(mpimg.imread('children.jpg'))
plt.subplot(224)
plt.imshow(mpimg.imread('workplace.jpg'))
```

```python
for idx, img in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
  plt.subplot(220 + idx + 1)
  plt.imshow(mpimg.imread(img))
```

```python
x = np.arange(-3,3,0.1)
x_sin = np.sin(x)
plt.plot(x,x_sin)  
```
