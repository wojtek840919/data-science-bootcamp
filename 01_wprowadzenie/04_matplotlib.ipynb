```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

```python
fig=plt.figure()
fig.show()
```

```python
_=plt.plot([0,1], [0,3])
```

```python
_=plt.plot([0,1,2], [0,3,2], label='line', color='green')
_=plt.legend()
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres')
```

```python
x=np.arange(-3,3,0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)
_=plt.plot(x, x_sin)
_=plt.plot(x, x_cos)
```

```python
plt.subplot(121)
plt.plot([0,1,2], [3,2,5])
plt.subplot(122)
plt.plot([0,1,2], [3,1,4], 'r--')
```

```python
fig=plt.figure(figsize=(4,4))
plt.subplot(131)
plt.plot([0,1,2], [3,2,5])
plt.subplot(132)
plt.plot([0,1,2], [3,1,4], 'r--')
plt.subplot(133)
plt.plot([0,1,2], [3,1,4], 'ro')
fig.show()
```

```python
plt.bar(x=[0,1,2,3], height=[1.2, 3.2, 5.2, 4.2])
```

```python
plt.bar(x=['a','b','c','d'], height=[1.2, 3.2, 5.2, 4.2])
```

```python
plt.barh(y=['a','b','c','d'], width=[1.2, 3.2, 5.2, 4.2])
```

```python
x1=np.random.randn(300)
x2=np.random.randn(300)
plt.scatter(x=x1, y=x2)
```

```python
x3=np.random.randn(300)*50
plt.scatter(x=x1, y=x2, s=x3)
```

```python
plt.scatter(x=x1, y=x2, s=x3, alpha=0.5)
```
